# Invert Color 

```python
from PIL import Image
```

```python
path = "imagen.png"
img = Image.open(path)
```

```python
width , height = img.size
```

```python
width , height
```

```python
print(img)
```

```python
img
```

```python
newimg = Image.new("RGB",(width,height),"white")
for i in range(1,width-1):
    for j in range(1,height-1):
        pixel = img.getpixel((i-1,j-1))
        red = pixel[0]
        green = pixel[1]
        blue = pixel[2]
        newimg.putpixel((i,j),(255-red,255-green,255-blue)) 
```
